# CNN (Convolution Neural Network)

### Convolution 개념 (데이터 관점에서)
1. Convolution의 수학적 개념
- 하나의 함수 f(x)와 또 다른 함수 g(x)가 있을 때, g(x)를 x에 대해 반전시키고 t 만큼 이동한 후 f(x)와 곱해서 모든 범위(-inf ~ inf)에서 적분하는 것을 의미한다.

2. Convolution의 데이터적 관점에서 개념
- 원본 데이터(=입력 데이터) f(x)에 대해 g(x-t)를 곱해서 원하는 출력 데이터로 변화를 준 후, 그 변화된 값들의 평균을 구한다는 의미이다.

- 참고
1. 데이터 관점에서 곱셈 : 원본 데이터에 대해 변화를 준다는 의미
1. 데이터 관점에서 덧셈 : 데이터의 평균을 구한다는 의미

### Convolution에서 시간의 전이 (time shift)
- 시간에 따른 이동 -> 곱셉 -> 덧셈

### 데이터 관점에서 Convolution 해석
- 시간의 흐름에 따라 데이터 g(x)가 이동하면서, 입력 데이터 f(x)를 평균적으로 얼마나 변화시키는지 나타내는 것
- 시간의 흐름에 따라 움직이는 데이터 g(x)에 의해, 입력 데이터 f(x)가 평균적으로 얼마나 변하는지 나타내는 것

### 일반적인 Neural Network(NN) 과 Convolutional Neural Network(CNN)의 차이점
CNNvsNN.PNG

- NN : 은닉층이 선형회귀와 활성화 함수 (ex. relu)로 이루어져 있다.
- CNN : 은닉층이 여러 개의 컨볼루션층과 마지막에 완전연결층으로 이루어져 있고 각각의 컨볼루션층은 컨볼루션, 활성화 함수, pooling을 포함하고 있다.

### CNN 컨볼루션층
1. conv : 입력 데이터와 가중치들의 집합체인 다양한 필터(filter)와의 컨볼루션 연산을 통해 입력데이터의 특징(feature)을 추출하는 역할을 수행함
- 예1) A1 * filter_1 + b2  =>> 입력데이터 A1의 특징(feature) 추출
- 예2) A2 * filter_2 + b3  =>> 입력데이터 A2의 특징(feature) 추출

2. relu : 활성화 함수 (입력으로 주어진 값이 0보다 크면 그대로 값을 출력하고, 0보다 작으면 0을 출력해줌)

3. pooling(풀링) : 입력 정보를 최대값, 최소값, 평균값 등으로 압축하여 데이터 연산량을 줄여주는 역할을 수행함 (일반적으로 max pooling이 가장 많이 사용됨)
- 예) conv 결과 = [ [15, 16], [6, 15] ] : 2 * 2 행렬
- max pooling = 16 : 1 * 1 행렬
- min pooling = 6 : 1 * 1 행렬
- average pooling = (15 + 16 + 6 + 15) / 4 = 13 : 1 * 1 행렬

### 내용 정리
- filter(필터) : 가중치들의 집합체
- stride(스트라이드) : 입력데이터와 필터를 컨볼루션 연산할 때 필터를 움직이는 간격
- Convolution(컨볼루션) : 필터를 일정간격(스트라이드)으로 이동해 가면서, 입력데이터와 대응하는 원소와 곱한 후 그 값들을 모두 더해주는 연산
- feature map(특징맵) : 컨볼루션 연산 결과에 바이어스를 더해준 값들

### Padding (패딩)
- 패딩이란 컨볼루션 연산을 수행하기 전에 입력 데이터 주변을 특정 값 (예를들면 0)으로 채우는 것을 말하며, 컨볼루션 연산에서 자주 사용되는 방법이다.
- 패딩을 하는 이유 : 컨볼루션 연산을 수행하면 데이터 크기(shape)가 줄어드는 단점을 방지하기 위해

## 컨볼루션 연산을 통한 출력 데이터 크기(shape) 계산
- 입력 데이터 크기 : (H, W)
- 필터 크기 : (FH, FW)
- 패딩 : P
- 스트라이드 : S
- 출력 데이터 크기 : (OH, OW)

1. OH = ( (H + 2P - FH) / S ) + 1
1. OW = ( (W + 2P - FW) / S ) + 1
